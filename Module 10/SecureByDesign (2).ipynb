```python
import csv
import hashlib
```

```python
# We assume:
# Segment is Username
# Country is Company

# Define User class with no privileges at __init__
class User:
    def __init__(self):
        
        self.username = "user1"
        self.access = ['read', 'write'] 


    def read_data(self): #This is function for reading data
        if 'read' in self.access:  #self.access means goes read and write 
            with open('sample (6).csv') as File:
                reader = csv.DictReader(File)
                for row in reader:
                    print(row)
                return "Completed"
        else:
            return "You cant read data"
        
    def write_data(self):       
        if 'write' in self.access:
            
            country = input("Enter the country name")
            Product = input("Enter the Product name:\n")
            
            with open('sample (6).csv', 'a') as File:
                writer = csv.writer(File)
                data = [country, Product]
                writer.writerow(data)
                print(data)
                return "Data Added Successfully"
        else:
            return "You cant write data"
```

```python
def main():
    user = User()
    while True:
        try:
            command = input("Enter Command",)
 
            if command == "read data":
                print(user.read_data())
 
            if command == "write data":
                print(user.write_data())
 
        except:
            pass

```

```python
main()
```
